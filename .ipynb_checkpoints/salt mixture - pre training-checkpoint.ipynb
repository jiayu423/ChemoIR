## PCA time

```python
from sklearn.decomposition import PCA 
```

```python
run1.shape, run2.shape, run3.shape
```

```python
train_all = np.vstack((run1, run2, run3))
```

```python
plt.plot(train_all[::10, :-2].T);
```

```python
n_comp = 20
salts = PCA(n_comp).fit(train_all[:, :-2]) 
plt.scatter([i+1 for i in range(n_comp)], np.cumsum(salts.explained_variance_ratio_)[:n_comp])
plt.xlabel('PCs')
plt.ylabel('Explained variance ratio')
```

```python
salts_pca = salts.transform(train_all[:, :-2])
```

```python
plt.scatter(salts_pca[:250, 0], salts_pca[:250, 1] ,label='1')
plt.scatter(salts_pca[250:250+225, 0], salts_pca[250:250+225, 1],label='2')
plt.scatter(salts_pca[250+225:, 0], salts_pca[250+225:, 1],label='3')
plt.legend()
```

```python
plt.plot(salts.components_[5, :])
plt.plot(train_all[120, :-2])
# plt.plot(salts.components_[1, :])
# plt.plot(salts.components_[2, :])
# plt.plot(salts.components_[3, :])
# plt.plot(salts.components_[4, :])
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
%matplotlib notebook
df1 = pd.DataFrame(data=salts_pca[:250, [0, 1, 2]], columns=['PC1', 'PC2', 'PC3'])
df2 = pd.DataFrame(data=salts_pca[250:250+225, [0, 1, 2]], columns=['PC1', 'PC2', 'PC3'])
df3 = pd.DataFrame(data=salts_pca[250+225:, [0, 1, 2]], columns=['PC1', 'PC2', 'PC3'])


fig = plt.figure(figsize=(8, 6))
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)
x1 = df1['PC1']
y1 = df1['PC2']
z1 = df1['PC3']

x2 = df2['PC1']
y2 = df2['PC2']
z2 = df2['PC3']

x3 = df3['PC1']
y3 = df3['PC2']
z3 = df3['PC3']

ax.scatter(x1, y1, z1, cmap='coolwarm')
ax.scatter(x2, y2, z2, cmap='coolwarm')
ax.scatter(x3, y3, z3, cmap='coolwarm')

ax.set_xlabel('pc1')
ax.set_ylabel('pc2')
ax.set_zlabel('pc3')
# plt.colorbar(cb, shrink=0.5, label='time')
plt.show()
```

```python
tot_data_cor = np.hstack((salts_pca[:, :6], train_conc_expend, train_all[:, -1].reshape(-1, 1)))
```

```python
import seaborn as sns

sns.set_theme(style="white")

# Generate a large random dataset
d = pd.DataFrame(data=tot_data_cor,
                 columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'CO3', 'NO3', 'SO4', 'NO2', 'T'])

# Compute the correlation matrix
corr = d.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)
```
