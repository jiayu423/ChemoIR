```python
na, nb = SVE.shape[1], SVE.shape[1]
pumpA = np.linspace(0, na, na)
pumpB = np.linspace(0, nb, nb)

plt.figure(figsize=(12, 10))
[AA,BB] = np.meshgrid(pumpA,pumpB)
plt.contourf(AA,BB,xErrMat)
plt.colorbar()
```

```python
RL = RobustLinear(xErrMat)
param_dict = {'lambda_':[1e-8, 1e-6, 1e-4, 1e-2, 1, 10, 100, 1000]}
gridCV = GridSearchCV(RL, param_grid=param_dict, scoring='neg_mean_absolute_error')
for i in range(4): 
    gridCV.fit(spec[:, :-2:2], conc[:, i]);
    print(gridCV.best_params_)
```

```python
RL = RobustLinear(xErrMat)
param_dict = {'lambda_':[50, 60, 70, 80, 90, 100, 160]}
gridCV = GridSearchCV(RL, param_grid=param_dict, scoring='neg_mean_absolute_error')
gridCV.fit(spec[:, :-2:2], conc[:, 1])
print(gridCV.best_params_)
```

```python
param_list = [0.01, 159, 10, 1000]
name_ = ['CO3', 'NO3', 'SO4', 'NO2']
MAE_RL = []
for i in range(4): 
    RL = RobustLinear(xErrMat, param_list[i], normalize=True)
    RL.fit(spec[:, :-2:2], conc[:, i])
    ytest = RL.predict(spec_test[:, :-2:2])
    MAE_RL.append(mean_absolute_error(conc_test[:, i], ytest))
    plt.scatter(conc_test[:, i], ytest, label=name_[i])
plt.legend()
plt.plot([-0.5, 8], [-0.5, 8])
```
